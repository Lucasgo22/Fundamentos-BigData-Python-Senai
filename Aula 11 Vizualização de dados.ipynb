```python
import matplotlib.pyplot as plt #Permite a geração de graficos
import numpy as np # Permite trabalhar com matrizes e vetores 
```

Primeiro Gráfico com matplotlib

```python
# Criando os Vetores (Conjunto de dados)
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

#plotar o grafico

plt.plot (x ,y)
```

### Modificando os tipos de marcadores

```python
plt.plot(x,y, marker= '*')
```

### alterando a cor da linha 

```python
plt.plot(x,y, marker= '*', c='#72DEEB')
```

### Mudando como os pontos são ligados e o tipo de linha

```python
plt.plot (
    x,y,                # Os dados do grafico
    marker = '*',       # Tipo de Marcador
    c = 'red',          # Cor da linha
    ds = 'steps',       # Forma de conexão dos pontos
    ls = '--',           # Estilo de linha
    mec = 'blue'        # Cor dos marcadores

)
```

```python
plt.plot (
    x,y,                # Os dados do grafico
    marker = '*',       # Tipo de Marcador
    c = 'red',          # Cor da linha
    ds = 'steps',       # Forma de conexão dos pontos
    ls = '--',           # Estilo de linha
    mec = 'blue'        # Cor dos marcadores

)
```

### Matplotlib - adicionando outros elemetos 

```python
#adiciona dados so grafico
plt.plot(x,y) 

#adicionar titulo ao eixo x

plt.xlabel('valores de x', fontsize =12)

# adicionando titulo ao eixo y
plt.ylabel('Valores de Y', fontsize = 12)

#adicionar titulo ao grafico
plt.title ('Titulo do grafico', fontsize = 18, c = 'Blue')

# alterando as cores dos valores dos eixos

plt.xticks (fontsize = 8, c = 'blue')
plt.yticks (fontsize = 8, c = 'red')
```

### Multiplas curvas no mecanismo gráfico


```python
import matplotlib.pyplot as plt
import numpy as np

#Retorna (linspace) valores igualmente espaçados em um intervalo
x = np.linspace (0,2*np.pi,100)
y = np.sin(x)
z = np.cos(x)
plt.plot (x,y)
plt.plot (x,z)
```

```python
# O comando plt.subplot permite criar grupo de grafico
# plt.subplot (linhas,colunas, posição elemento)

plt.subplot(2,1,1) # priemiro grafico
plt.plot(x,y) #conjunto de dados primeiro grafico
plt.title('subplots')


plt.subplot(2,1,2) #Segundo grafico
plt.plot(x,z) # conjunto de dados do segundo grafico
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('subplots')
plt.subplot(1,2,2)
plt.plot(x,z)
```

### Seaborn – Gráficos mais elaborados
Seaborn é uma biblioteca baseada em Matplolib que traz uma interface mais atraente para os gráficos,
além de facilitar a modificação de parâmetros default do Matplotlib e permite a utilização de dataframes.

```python
# o ALIAS (apelido) da biblioteca seaborn fica com sns
import seaborn as sns
import numpy as np

# Retorna (Linspace) valores igualmente espaçados em um intervalo
x = np.linspace (0,2*np.pi,100) 
y = np.sin(x) 
```

```python
sns.get_dataset_names()
```

```python

iris = sns.load_dataset('iris')
iris.head()
```

### Seaborn - ajustando os grafico

```python
import matplotlib.pyplot as plt

sns.set()
plt.plot(x,y)
```

```python
#alterando o tema para claro

sns.set_style('whitegrid')
plt.plot(x,y)
```

```python
#alterando o tema para escuro

sns.set_style('dark')
plt.plot(x,y)
```

### Seaborn - dados categóricos

```python
#Grafico de barras

sns.barplot(x='species',y='petal_length',data=iris)
```
