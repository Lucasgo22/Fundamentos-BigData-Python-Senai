### Exercício 01 - Escreva um programa Python para desenhar uma linha com rótulo
adequado no eixo x, eixo y e um título.

```python
# importando a biblioteca
import matplotlib.pyplot as plt #Permite a geração de graficos
import numpy as np # Permite trabalhar com matrizes e vetores 


# Gerar os daos para inserir no grafico

x = range (1,50)
y = [value * 3 for value in x]

print ('valores de x')
print (range (1,50))
print ('valores de y (tres vezes x)')
print (y)

#plotar um grafico de linha

plt.plot(x,y)

# adicionar um titulo ao grafico
plt.title('Draw a line', fontsize = 18 , c = 'red')

# adicionar titulo ao eixos
plt.xlabel ('valores de x')
plt.ylabel('valores de y')
```

Exercício 02 - Escreva um programa Python para desenhar uma linha usando
determinados valores de eixo com rótulo adequado no eixo x, eixo y e um título.

```python
#adicionando a biblioteca
import matplotlib.pyplot as plt 
import numpy as np

#adicionanto os valores de x e de y
x = np.array ([1,2,3])
y = np.array ([2,4,1])

#plotanto o grafico
plt.plot(x,y)

plt.title('Sample graph', fontsize = 20 , c = 'red')
plt.xlabel('X - axis', c ='blue')
plt.ylabel('Y - axis', c = 'pink')
```
